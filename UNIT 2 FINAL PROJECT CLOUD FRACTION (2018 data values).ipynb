```python
#import needed libraries
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
csvFile = open("CloudFraction1_2018.csv", 'r')
##csvFiletwo = open("CloudFraction1_2002.csv", 'r')
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
print(len(data_lists[0]))
```

```python
width = len(data_lists[0]) ##720 
height = len(data_lists)
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go
#darker colours = lower values
for r, row in enumerate(data_lists):
    for c, value in enumerate(row): #ADDS the ability to keep track of one more thing; when i is 1, C = 0; keeps track of where you are in the list in terms of positioning
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0:
            grid[r, c] = [2, 25, 46] #darkest blue
        elif float(value) <= 0.1:
            grid[r, c] = [3, 41, 77]
        elif float(value) <= 0.2:
            grid[r, c] = [8, 76, 138]
        elif float(value) <= 0.3:
            grid[r, c] = [28, 99, 163]
        elif float(value) <= 0.4:
            grid[r, c] = [6, 99, 184]
        elif float(value) <= 0.5:
            grid[r, c] = [38, 124, 201]
        elif float(value) <= 0.6:
            grid[r, c] = [7, 119, 222]
        elif float(value) <= 0.7:
            grid[r, c] = [10, 131, 242]   
        elif float(value) <= 0.8:
            grid[r, c] = [56, 160, 255]
        elif float(value) <= 0.9:
            grid[r, c] = [124, 191, 252]
        elif float(value) <= 1.0:
            grid[r, c] = [159, 208, 252] #really light, almost white
        else:
            grid[r, c] = [0,0,0]
```

```python
print(max(all_values))
print(min(all_values))
```

```python
plt.title("Cloud Fracturing 2018")
plt.imshow(grid)
plt.axis('off')
##plt.savefig("Cloud fracturing.png",dpi = 1000)
plt.show()
```
